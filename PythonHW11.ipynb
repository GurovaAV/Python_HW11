### Исследование функции f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

#### Импорт библиотек

```python
from sympy import*
from sympy.plotting import plot
from matplotlib import*
```

#### Построение графика функции

```python
x = Symbol('x')
f = -12*x*x*x*x*sin(cos(x))-18*x*x*x+5*x*x+10*x-30
plot(f,(x,-70,70),line_color='orange')
```

#### Определение корней функции

_Судя по графику, данная функция имеет бесконечное множество корней_

#### Нахождение производной функции

```python
print(f'Производная функции: {f.diff(x)}')
diff = f.diff(x)
plot(diff,(x,-70,70),line_color='green',title='Производная функции diff(x)')
```

#### Интервалы возрастания и убывания функции

Функция убывает при diff(x) < 0 

Функция возрастает при diff(x) > 0

#### Определение вершины графика функции

~Вообще вершины вроде есть только у парабол, но возможно здесь речь идет о пересечении с осью Y~

```python
print(f'Точка пересечения графика функции с осью Y [0 ,{f.subs(x, 0)}].')
```
